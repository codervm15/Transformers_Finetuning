```python
import tensorflow as tf
```

```python
(X_train, X_test), (Y_train, Y_test) = tf.keras.datasets.mnist.load_data()
```

```python
print(X_train.shape)
print(Y_train.shape)

print(X_test.shape)
print(Y_test.shape)
```

```python
## Initializing constants

input_width = 28      # Input image width
input_height = 28     # Input image height
input_channels = 1    # Input channels (means depth)
input_pixels = 784    # Input features

n_conv1 = 32          # No. of units in convolution layer 1
n_conv2 = 64          # No. of units in Conv layer 2
stride_conv1 = 1      # Stride value for conv layer 1
stride_conv2 = 1      # stride value for conv layer 2
conv1_k = 5           # 5x5 filter value for conv1
conv2_k = 5           # 5x5 filter value for conv2
max_pool1_k = 2         # 2x2 filter value for max pooling layer 1
max_pool2_k = 2         # 2x2 filter value for max pooling layer 2

n_hidden = 1024       # No. of units in hidden layer
n_out = 10            # No. of units in output layer
input_size_to_hidden = (input_width//(max_pool1_k*max_pool2_k)) * (input_height//(max_pool1_k*max_pool2_k)) * n_conv2
```

```python
## Defining weights and biases

weights = {
    "wc1" : tf.Variable(tf.random.normal([conv1_k, conv1_k, input_channels, n_conv1])),
    "wc2" : tf.Variable(tf.random.normal([conv2_k, conv2_k, n_conv1, n_conv2])),
    "wh1" : tf.Variable(tf.random.normal([input_size_to_hidden, n_hidden])),
    "w0" : tf.Variable(tf.random.normal([n_hidden, n_out])),
}

biases = {
    "bc1" : tf.Variable(tf.random.normal([n_conv1])),
    "bc2" : tf.Variable(tf.random.normal([n_conv2])),
    "bh1" : tf.Variable(tf.random.normal([n_hidden])),
    "bo" : tf.Variable(tf.random.normal([n_out]))
}
```

```python
def conv(x, weights, bias, strides=1):
  out = tf.nn.conv2d(x, weights, padding="SAME", strides=[1,strides, strides, 1])
  out = tf.nn.bias_add(out, bias)
  out = tf.nn.relu(out)

  return out

def maxpooling(x, k=2):
  return tf.nn.max_pool(x, padding="SAME", ksize=[1,k,k,1], strides=[1,k,k,1])
```

```python
def cnn(x , weights, biases):

  x = tf.reshape(x , shape = [-1, input_height, input_width, input_channels])

  conv1 = conv(x, weights['wc1'], biases['bc1'], stride_conv1)
  conv1_pool = maxpooling(conv1, max_pool1_k)

  conv2 = conv(x, weights['wc2'], biases['bc2'], stride_conv2)
  conv2_pool = maxpooling(conv2, max_pool2_k)

  hidden_input = tf.reshape(conv2_pool, shape = [-1, input_size_to_hidden])
  hidden_output_before_activation = tf.add(tf.matmul(hidden_input, weights["wh1"]), biases["bh1"])

  hidden_output = tf.nn.relu(hidden_output_before_activation)

  output = tf.add(tf.matmul(hidden_output, weights["wo"]), biases["bo"])
```

```python
x = tf.compat.v1.placeholder("float", [None, input_pixels])
y = tf.compat.v1.placeholder(tf.int32, [None, n_out])

pred = cnn(x, weights, biases)
```

```python
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=pred, labels=y))
```

```python
optimizer = tf.train.AdamOptimizer(learning_rate = 0.01)
optimize = optimizer.minimize(cost)
```
