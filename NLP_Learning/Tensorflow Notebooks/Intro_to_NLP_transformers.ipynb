```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation" , model = "gpt2")


generator(
    "My name is Vikrant. And my learning objectives are",
    max_length = 100,
    num_return_sequences = 3,
)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("Smishing is a phishing cybersecurity attack carried out over mobile text messaging, also known as SMS phishing.")
```
