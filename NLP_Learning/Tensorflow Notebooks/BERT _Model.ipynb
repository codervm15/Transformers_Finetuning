```python
from transformers import AutoTokenizer

checkpoint1 = "bert-base-cased"
checkpoint2 = "bert-base-uncased"

tokenizer1 = AutoTokenizer.from_pretrained(checkpoint1)
tokenoizer = AutoTokenizer.from_pretrained(checkpoint2)
```

```python
raw_text = ["I want to learn [MASK]"]

inputs = tokenizer1(raw_text, return_tensors="tf")
```

```python
print(inputs)
```

```python
from transformers import TFAutoModel

model1 = TFAutoModel.from_pretrained(checkpoint1)
```

```python
outputs = model1(inputs)
print(outputs)
```
